## Exercise 3: Slicing

---

np.arange(start, stop) generates a sequence of numbers starting from start up to, but not including, stop.
In our example, np.arange(1, 101) produces numbers from 1 to 100 because the start is 1 and the stop value 101 is exclusive

````python
arr = np.arange(1, 101)
print(arr)
````

````python
#array of all ods
odd_arr = arr[arr % 2 != 0]
print(odd_arr)
````

````python
#array of all even
even_arr = arr [arr % 2 == 0]
print(even_arr)
````

````python
#array of all even arrays in reverse
reverse_even_arr = arr [arr % 2 == 0][::-1]
print(reverse_even_arr)
````

````python
#Set every 3rd element (starting from the second) to 0
arr_copy = arr.copy() # keeping original intact
arr_copy[1::3] = 0
print(arr_copy)
````

- arr % 2 != 0 selects odd numbers.

- [::-1] reverses an array.

- 1::3 selects every 3rd element starting from index 1.

## Exercise 4: Random

NumPy lets us generate random numbers in different ways. **rand()** gives random decimals between 0 and 1, **randint(a, b)** gives whole numbers between a and b-1, **randn()** generates numbers from a normal distribution (centered around 0), and **choice()** randomly picks elements from a list we provide. This makes it easy to simulate data or test algorithms

## Exercise 6: Broadcasting and Slicing

The goal of this exercise is to learn how to access and manipulate values of n-dimensional arrays efficiently using slicing and broadcasting.




````python
### Part A: Create a 9×9 array with borders and a pattern
arr = np.ones((9, 9), dtype=np.int8)
print(arr)
````

````python
#Step 2: Use slicing to insert zeros inside (keeping border = 1)
ar = arr[1:-1, 1:-1] = 0
#Add the inner pattern
arr[2, 2:7] = 1
arr[3, [2, 6]] = 1
arr[4, [2, 4, 6]] = 1
arr[5, [2, 6]] = 1
arr[6, 2:7] = 1

print(arr)
````

````python
#Using broadcasting create an output matrix based on the following two arrays:
array_1 = np.array([1, 2, 3, 4, 5], dtype=np.int8)
array_2 = np.array([1, 2, 3], dtype=np.int8)

#Multiply using broadcasting
result = array_1[:, np.newaxis] * array_2   # shape (5,3)
print(result)

````

## Exercise 7: Handling Missing Data with np.where

Here we handled NaN (missing values) in a dataset of student grades without using loops.

✅ Problem

We have a (10 × 2) array of grades from two exams. Some students missed the first exam, so their score is stored as NaN.

We want to create a third column:

- If the first exam grade is available → use it.

- If it’s missing (NaN) → use the second exam grade.

````python
import numpy as np

# Step 1: Generate dataset
generator = np.random.default_rng(123)
grades = np.round(generator.uniform(low=0.0, high=10.0, size=(10, 2)))
grades[[1, 2, 5, 7], [0, 0, 0, 0]] = np.nan
print("Original grades:\n", grades)

# Step 2: Create third column using np.where
third_col = np.where(np.isnan(grades[:, 0]), grades[:, 1], grades[:, 0])

# Step 3: Concatenate third column to grades
grades_final = np.column_stack((grades, third_col))

print("\nFinal grades with 3rd column:\n", grades_final)
````

## Exercise 8: Wine

````python
import numpy as np

# Step 1: URL of the CSV
url = "https://learn.zone01kisumu.ke/api/content/root/public/subjects/ai/numpy/data/winequality-red.csv"

# Step 2: Load data with genfromtxt
data = np.genfromtxt(url, 
                     delimiter=";", 
                     skip_header=1, 
                     dtype=np.float32)

# Step 3: Initial checks
print("Shape of data:", data.shape)        # how many rows and columns
print("First 5 rows:\n", data[:5])         # show first 5 rows
print("Has any NaNs:", np.any(np.isnan(data)))
print("Memory usage (bytes):", data.nbytes)



# If the bytes count ≠ 76800, check: dtype and shape
````

````python
#2  Extract the 2nd, 7th, and 12th rows
selected_rows = data[[1, 6, 11], :]
print(selected_rows)

# Remove NaNs from each row
cleaned_rows = [row[~np.isnan(row)] for row in selected_rows]
cleaned_rows = np.array(cleaned_rows, dtype=np.float32)
print(cleaned_rows)


````

````python
# 3 determine if alcohol percentage greater than 20%
alcohol = data[:, 10]   # get all rows, column 10 (alcohol %)

has_high_alcohol = np.any(alcohol > 20)
print(has_high_alcohol)

````

````python
# 4 calculate % alcohol of all wine
#Step 1: Extract the alcohol column
alcohol = data[:, 10]

#Step 2: Exclude NaN values (just in case)
alcohol_no_nan = alcohol[~np.isnan(alcohol)]

#Step 3: Compute mean
avg_alcohol = np.mean(alcohol_no_nan)
print("Average alcohol %:", avg_alcohol)

````

````python
# 5: Statistical measures for the pH values.

#Step 1: Extract the pH column
ph = data[:, 8]

#Step 2: Remove NaN values (if any)
ph_no_nan = ph[~np.isnan(ph)]

#Step 3: Compute the statistics
ph_min = np.min(ph_no_nan)
ph_max = np.max(ph_no_nan)
ph_25 = np.percentile(ph_no_nan, 25)
ph_50 = np.percentile(ph_no_nan, 50)   # same as median
ph_75 = np.percentile(ph_no_nan, 75)
ph_mean = np.mean(ph_no_nan)

print("pH Statistics:")
print("Min:", ph_min)
print("Max:", ph_max)
print("25th percentile:", ph_25)
print("50th percentile (median):", ph_50)
print("75th percentile:", ph_75)
print("Mean:", ph_mean)
````

````python
#6: Average quality for wines with the 20% least sulphate content.
#Step 1: Extract the sulphates and quality columns

sulphates = data[:, 9]
quality = data[:, 11]

sulphates_20 = np.percentile(sulphates, 20)
print("20th percentile of sulphates:", sulphates_20)

#Create a boolean mask for wines with sulphates below this threshold
mask = sulphates < sulphates_20

low_sulphate_quality_mean = np.mean(quality[mask])
print("Average quality (20% least sulphates):", low_sulphate_quality_mean)

````

````python
#7: Mean of all variables for best and worst quality wines
quality = data[:, 11]

best_quality = np.max(quality)
worst_quality = np.min(quality)

print("Best quality:", best_quality)
print("Worst quality:", worst_quality)

mask_best = quality == best_quality
mask_worst = quality == worst_quality

mean_best = np.mean(data[mask_best], axis=0)
mean_worst = np.mean(data[mask_worst], axis=0)

print("Mean values for best quality wines:\n", mean_best)
print("\nMean values for worst quality wines:\n", mean_worst)

````

## Exercise 9: Football tournament

````python
import numpy as np
import itertools

# Step 1: Load forecast matrix (10x10)
url = "https://learn.zone01kisumu.ke/api/content/root/public/subjects/ai/numpy/data/model_forecasts.txt"
matrix = np.genfromtxt(url, delimiter=',')

# Step 2: Create list of 10 teams
teams = np.arange(10)

# Step 3: Generate all possible team pairs
all_pairs = list(itertools.combinations(teams, 2))

# Step 4: Function to calculate sum of squared differences for a given set of matches
def pairing_cost(pairs):
    return sum(matrix[t1, t2]**2 for t1, t2 in pairs)

# Step 5: Generate candidate match combinations
# We pick 5 pairs that do not overlap (i.e., no team appears twice)
# Use a simple method with itertools.permutations for illustration (may not be fully optimal)
def generate_non_overlapping_matches(pairs):
    # Try all combinations of 5 pairs
    for candidate in itertools.combinations(pairs, 5):
        # Flatten all teams in candidate pairs
        flat = [t for pair in candidate for t in pair]
        # Check if all teams are unique
        if len(set(flat)) == 10:
            yield candidate

# Step 6: Find the combination with minimum sum of squared differences
best_cost = float('inf')
best_matches = None

for candidate in generate_non_overlapping_matches(all_pairs):
    cost = pairing_cost(candidate)
    if cost < best_cost:
        best_cost = cost
        best_matches = candidate

# Step 7: Format result as 2x5 array
result = np.array(best_matches).T
print("Best matches (Team1 vs Team2):")
print(result)
````
